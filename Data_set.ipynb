<h2><font size=4>Data set</h2>

<p>Data set which will be using in the project is collision records in Seattle, USA from SDOT Traffic Management Division, Traffic Records Group. Which contains all types of collisions. Collisions will display at the intersection or mid-block of a segment from 2004 to Present.</p>
<p>The data set will be used to check correlation of the incident with the weather, road condition, light condition, speeding, drugs and inattention.</p>
<p>Weather, Road condition and Light conditions are natural factors which can be notified to the drivers. Furthermore, speeding, drugs and inattention is involved for collisions, We should look for more socialistic way to prevent or reduce such incidents.
</p>
